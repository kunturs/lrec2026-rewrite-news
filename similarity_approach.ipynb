```python
from sentence_transformers import SentenceTransformer, util
import torch
```

```python
model = SentenceTransformer('sentence-transformers/distiluse-base-multilingual-cased-v2', 
                            device = 0) # optional define GPU
```

```python
# Find similar sentences. Treshold 0.6 
def find_sim(tensor1, tensor2, threshold=0.6):
    # Ensure tensors
    if not isinstance(tensor1, torch.Tensor):
        tensor1 = torch.tensor(tensor1)
    if not isinstance(tensor2, torch.Tensor):
        tensor2 = torch.tensor(tensor2)

    device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu') # optional, define GPU
    tensor1 = tensor1.to(device)
    tensor2 = tensor2.to(device)

    sim = util.pytorch_cos_sim(tensor1, tensor2)
    return sim.item() > threshold
```

```python
device = torch.device('cuda:1' if torch.cuda.is_available() else 'cpu') # optional, define GPU

sentence1 = "According to the United States Census Bureau, the city has a total area of , all of it land."
sentence2 = "According to the United States Census Bureau, the town has a total area of , of which is land and (2.42 ) is water."
tensor1 = model.encode(' '.join(sentence1), convert_to_tensor=True)
tensor2 = model.encode(' '.join(sentence2), convert_to_tensor=True)
    
tensor1 = tensor1.to(device) # optional
tensor2 = tensor2.to(device) # optional

sim = find_sim(tensor1, tensor2, threshold=0.6)
```
